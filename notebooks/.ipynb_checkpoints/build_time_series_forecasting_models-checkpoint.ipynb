```python
# Cell 1: Clean Imports (No TensorFlow)

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge
import warnings
warnings.filterwarnings('ignore')

# Try pmdarima for auto_arima
try:
    from pmdarima import auto_arima
    PMDARIMA_AVAILABLE = True
    print("✅ pmdarima available!")
except ImportError:
    PMDARIMA_AVAILABLE = False
    print("⚠️ pmdarima not installed - using manual ARIMA selection")
    print("   To install: pip install pmdarima")

# Plotting settings
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.4f' % x)

print("\n" + "="*60)
print("✅ All libraries imported successfully!")
print("   ✅ Using: ARIMA, Random Forest, MLP, Ridge")
print("   ❌ LSTM skipped (TensorFlow not available)")
print("="*60)
```

```python
# Cell 2: Load Cleaned Data

# Load cleaned Tesla data
tsla_df = pd.read_csv('../data/processed/TSLA_cleaned.csv', index_col=0, parse_dates=True)
print(f"📊 Tesla data loaded: {len(tsla_df)} rows")
print(f"   Date range: {tsla_df.index.min()} to {tsla_df.index.max()}")
display(tsla_df.head())

# Focus on closing prices for forecasting
tsla_close = tsla_df['Close'].copy()
print(f"\n📈 Closing price data shape: {tsla_close.shape}")
display(tsla_close.head())
```

```python
# Cell 3: Complete Working Split

print("="*60)
print("🔄 Preparing Train-Test Split")
print("="*60)

# Step 1: Remove timezone from index
tsla_close_clean = tsla_close.copy()
if hasattr(tsla_close_clean.index, 'tz') and tsla_close_clean.index.tz is not None:
    tsla_close_clean.index = tsla_close_clean.index.tz_localize(None)
    print("✅ Removed timezone from index")

# Step 2: Check if dates exist in data
date_to_check = '2024-12-31'
if date_to_check in tsla_close_clean.index:
    print(f"✅ Date {date_to_check} exists in data")
    # Use date-based split
    train_data = tsla_close_clean.loc[:'2024-12-31']
    test_data = tsla_close_clean.loc['2025-01-01':]
    split_method = "date-based (2024-12-31)"
else:
    print(f"⚠️ Date {date_to_check} not found in data")
    # Use positional split
    split_idx = int(len(tsla_close_clean) * 0.8)
    train_data = tsla_close_clean.iloc[:split_idx]
    test_data = tsla_close_clean.iloc[split_idx:]
    split_method = "80-20 positional"

print(f"\n📊 Data Split Summary (Method: {split_method}):")
print("="*60)
print(f"Training period: {train_data.index.min()} to {train_data.index.max()}")
print(f"   Number of days: {len(train_data)} ({len(train_data)/len(tsla_close)*100:.1f}%)")
print(f"Test period: {test_data.index.min()} to {test_data.index.max()}")
print(f"   Number of days: {len(test_data)} ({len(test_data)/len(tsla_close)*100:.1f}%)")

# Visualize
fig, ax = plt.subplots(figsize=(15, 6))
ax.plot(train_data.index, train_data, label='Training Data', color='blue', linewidth=1.5)
ax.plot(test_data.index, test_data, label='Test Data', color='red', linewidth=1.5)
ax.axvline(x=train_data.index[-1], color='green', linestyle='--', 
           linewidth=2, label=f'Train-Test Split: {train_data.index[-1].date()}')
ax.set_title('TSLA Stock Price - Train/Test Split', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n✅ Split complete!")
print(f"   Training: {len(train_data)} days")
print(f"   Testing: {len(test_data)} days")

# Store the cleaned data for future use
tsla_close = tsla_close_clean
```

```python
# Cell 4: Stationarity Test

def adf_test_stationarity(series, title):
    """
    Perform ADF test and check stationarity
    """
    result = adfuller(series.dropna())
    print(f"\n📊 ADF Test Results - {title}")
    print("="*60)
    print(f"ADF Statistic: {result[0]:.6f}")
    print(f"p-value: {result[1]:.6f}")
    print(f"Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value:.6f}")
    
    if result[1] <= 0.05:
        print(f"\n✅ {title} is STATIONARY (p-value < 0.05)")
        return True
    else:
        print(f"\n❌ {title} is NON-STATIONARY (p-value >= 0.05)")
        print(f"   Differencing required")
        return False

# Test original prices
adf_test_stationarity(train_data, "Original Prices")

# Test returns (first difference)
train_returns = train_data.diff().dropna()
adf_test_stationarity(train_returns, "First Differences (Returns)")

# Determine d for ARIMA
d = 1  # First difference is needed
print(f"\n✅ Recommended differencing order (d): {d}")
```

```python
# Cell 5: ACF and PACF Analysis

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

# ACF plot for differenced series (returns)
plot_acf(train_returns, ax=ax1, lags=40)
ax1.set_title('Autocorrelation Function (ACF) - TSLA Returns', fontsize=12)
ax1.grid(True, alpha=0.3)

# PACF plot for differenced series
plot_pacf(train_returns, ax=ax2, lags=40)
ax2.set_title('Partial Autocorrelation Function (PACF) - TSLA Returns', fontsize=12)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📊 ACF/PACF Interpretation:")
print("="*60)
print("🔹 ACF shows significant lags indicating MA components")
print("🔹 PACF shows significant lags indicating AR components")
print("🔹 Based on the plots, we can estimate p and q parameters")
```

```python
# Cell 6a: ARIMA Parameter Selection

# Option 1: Using auto_arima (if available)
if PMDARIMA_AVAILABLE:
    print("🔍 Using auto_arima to find optimal parameters...")
    print("   This may take a few minutes...")
    
    try:
        auto_model = auto_arima(
            train_data, 
            start_p=0, max_p=5,
            start_d=1, max_d=2,  # We know d=1 from ADF test
            start_q=0, max_q=5,
            seasonal=False,
            stepwise=True,
            trace=True,
            error_action='ignore',
            suppress_warnings=True,
            n_fits=20
        )
        
        best_p = auto_model.order[0]
        best_d = auto_model.order[1]
        best_q = auto_model.order[2]
        
        print(f"\n✅ Best ARIMA parameters found:")
        print(f"   p = {best_p}, d = {best_d}, q = {best_q}")
        print(f"   AIC: {auto_model.aic():.2f}")
        
    except Exception as e:
        print(f"⚠️ auto_arima failed: {e}")
        print("Using manual parameter selection...")
        PMDARIMA_AVAILABLE = False

# Option 2: Manual parameter selection (if auto_arima not available)
if not PMDARIMA_AVAILABLE:
    print("🔍 Performing manual ARIMA parameter selection...")
    
    # Test different p, d, q combinations
    p_values = range(0, 4)
    d_values = [1]  # We know d=1
    q_values = range(0, 4)
    
    best_score = float('inf')
    best_params = None
    best_model = None
    
    print("\nTesting parameter combinations...")
    for p in p_values:
        for d in d_values:
            for q in q_values:
                try:
                    model = ARIMA(train_data, order=(p, d, q))
                    model_fit = model.fit()
                    aic = model_fit.aic
                    
                    if aic < best_score:
                        best_score = aic
                        best_params = (p, d, q)
                        best_model = model_fit
                    
                    print(f"   ARIMA({p},{d},{q}) - AIC: {aic:.2f}")
                except:
                    continue
    
    best_p, best_d, best_q = best_params
    print(f"\n✅ Best ARIMA parameters found:")
    print(f"   p = {best_p}, d = {best_d}, q = {best_q}")
    print(f"   Best AIC: {best_score:.2f}")
```

```python
# Cell 6b: Fit ARIMA Model

print("📊 Fitting ARIMA Model...")
print(f"   Parameters: ARIMA({best_p}, {best_d}, {best_q})")

# Fit ARIMA model
arima_model = ARIMA(train_data, order=(best_p, best_d, best_q))
arima_fit = arima_model.fit()

print("\n✅ Model Summary:")
print("="*60)
print(arima_fit.summary())

# Save model summary for later
arima_aic = arima_fit.aic
print(f"\n📊 Model AIC: {arima_aic:.2f}")
```

```python
# Cell 6c: ARIMA Forecast

# Generate forecasts for test period
forecast_steps = len(test_data)
arima_forecast = arima_fit.forecast(steps=forecast_steps)
arima_forecast_index = test_data.index

# Create DataFrame with predictions
arima_predictions = pd.DataFrame({
    'Actual': test_data.values,
    'Predicted': arima_forecast.values
}, index=arima_forecast_index)

print("📊 ARIMA Predictions (First 10 days):")
display(arima_predictions.head(10))

# Visualize predictions
fig, ax = plt.subplots(figsize=(15, 6))

# Plot training data
ax.plot(train_data.index, train_data, label='Training Data', color='blue', alpha=0.7)

# Plot actual test data
ax.plot(test_data.index, test_data, label='Actual Test Data', color='green', linewidth=2)

# Plot ARIMA predictions
ax.plot(arima_forecast_index, arima_forecast, label='ARIMA Predictions', 
        color='red', linewidth=2, linestyle='--')

# Add confidence intervals (optional - if available)
try:
    forecast_ci = arima_fit.get_forecast(steps=forecast_steps).conf_int()
    ax.fill_between(arima_forecast_index, 
                    forecast_ci.iloc[:, 0], 
                    forecast_ci.iloc[:, 1],
                    color='red', alpha=0.1, label='95% CI')
except:
    pass

ax.axvline(x=pd.Timestamp('2024-12-31'), color='black', linestyle='--', 
           linewidth=1, alpha=0.5, label='Train-Test Split')
ax.set_title('ARIMA Model Predictions vs Actual', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cell 7: SARIMA Model

# Check if data has seasonal patterns
# First, visualize seasonal decomposition
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform decomposition on training data
decomposition = seasonal_decompose(train_data, model='additive', period=252)  # 252 trading days in a year

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 12))
decomposition.observed.plot(ax=ax1, title='Original', color='blue')
decomposition.trend.plot(ax=ax2, title='Trend', color='green')
decomposition.seasonal.plot(ax=ax3, title='Seasonal', color='orange')
decomposition.resid.plot(ax=ax4, title='Residual', color='red')
plt.tight_layout()
plt.show()

# Try SARIMA if seasonal patterns exist
print("📊 SARIMA Model (with seasonality)")
print("="*60)

# Determine seasonal parameters (m = number of periods)
m = 252  # annual seasonality (252 trading days)

# Use auto_arima with seasonality if available
if PMDARIMA_AVAILABLE:
    try:
        print("🔍 Using auto_arima with seasonality...")
        sarima_model = auto_arima(
            train_data,
            start_p=0, max_p=3,
            start_d=1, max_d=2,
            start_q=0, max_q=3,
            seasonal=True,
            m=m,
            start_P=0, max_P=2,
            start_D=0, max_D=1,
            start_Q=0, max_Q=2,
            stepwise=True,
            trace=True,
            error_action='ignore',
            suppress_warnings=True
        )
        
        # Get parameters
        p, d, q = sarima_model.order
        P, D, Q, m = sarima_model.seasonal_order
        
        print(f"\n✅ Best SARIMA parameters:")
        print(f"   (p,d,q) = ({p},{d},{q})")
        print(f"   (P,D,Q,m) = ({P},{D},{Q},{m})")
        
        # Fit SARIMA model
        sarima_fit = SARIMAX(train_data, 
                            order=(p, d, q),
                            seasonal_order=(P, D, Q, m),
                            enforce_stationarity=False,
                            enforce_invertibility=False)
        sarima_fit = sarima_fit.fit(disp=False)
        
        print("\n✅ SARIMA Model Summary:")
        print(sarima_fit.summary())
        
        # Generate predictions
        sarima_forecast = sarima_fit.forecast(steps=len(test_data))
        
        # Store for evaluation
        sarima_predictions = pd.DataFrame({
            'Actual': test_data.values,
            'Predicted': sarima_forecast.values
        }, index=test_data.index)
        
        print("\n📊 SARIMA Predictions (First 10 days):")
        display(sarima_predictions.head(10))
        
        # Visualize
        fig, ax = plt.subplots(figsize=(15, 6))
        ax.plot(train_data.index, train_data, label='Training Data', color='blue', alpha=0.7)
        ax.plot(test_data.index, test_data, label='Actual Test Data', color='green', linewidth=2)
        ax.plot(sarima_forecast.index, sarima_forecast, label='SARIMA Predictions', 
                color='purple', linewidth=2, linestyle='--')
        ax.axvline(x=pd.Timestamp('2024-12-31'), color='black', linestyle='--', 
                   linewidth=1, alpha=0.5, label='Train-Test Split')
        ax.set_title('SARIMA Model Predictions vs Actual', fontsize=14)
        ax.set_xlabel('Date')
        ax.set_ylabel('Price (USD)')
        ax.legend()
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        SARIMA_AVAILABLE = True
        
    except Exception as e:
        print(f"⚠️ SARIMA failed: {e}")
        SARIMA_AVAILABLE = False
else:
    print("⚠️ pmdarima not available - skipping SARIMA")
    SARIMA_AVAILABLE = False
```

```python
# Cell 8a: Complete Data Preparation (Self-Contained)

print("🚀 STARTING DATA PREPARATION...")
print("="*60)

# Check if data exists
try:
    print(f"✅ train_data: {len(train_data)} rows")
    print(f"✅ test_data: {len(test_data)} rows")
except NameError:
    print("❌ Error: train_data/test_data not found!")
    print("   Please run Cell 3 first (Train-Test Split)")
    raise

# Create scaler and scale data
print("\n📊 Scaling data...")
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data.values.reshape(-1, 1))
test_scaled = scaler.transform(test_data.values.reshape(-1, 1))
print(f"   ✅ Train scaled: {train_scaled.shape}")
print(f"   ✅ Test scaled: {test_scaled.shape}")

# Define window size
window_size = 60
print(f"   ✅ Window size: {window_size}")

# Create sequences function
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

# Create sequences
print("\n📊 Creating sequences...")
X_train, y_train = create_sequences(train_scaled, window_size)
X_test, y_test = create_sequences(test_scaled, window_size)
print(f"   ✅ X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"   ✅ X_test: {X_test.shape}, y_test: {y_test.shape}")

# Reshape for LSTM
print("\n📊 Reshaping for LSTM...")
X_train_lstm = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_lstm = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
print(f"   ✅ X_train_lstm: {X_train_lstm.shape}")
print(f"   ✅ X_test_lstm: {X_test_lstm.shape}")

print("\n" + "="*60)
print("✅ LSTM Data Preparation Complete!")
print("="*60)
print(f"Training sequences: {X_train_lstm.shape[0]}")
print(f"Training shape: {X_train_lstm.shape}")
print(f"Test sequences: {X_test_lstm.shape[0]}")
print(f"Test shape: {X_test_lstm.shape}")
print(f"Window size: {window_size} days")
print(f"Features: 1 (Closing Price)")
print("="*60)
print("✅ Data preparation complete! Ready for LSTM model building.")
```

```python
# Cell 8b: Train ML Models (If not already done)

print("🚀 Training ML Models...")
print("="*60)

from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge

# Prepare data (if not already done)
window_size = 60
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data.values.reshape(-1, 1))
test_scaled = scaler.transform(test_data.values.reshape(-1, 1))

def create_sequences(data, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_scaled, window_size)
X_test, y_test = create_sequences(test_scaled, window_size)

# 1. Random Forest
print("📌 Training Random Forest...")
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred_scaled = rf_model.predict(X_test)
rf_pred = scaler.inverse_transform(rf_pred_scaled.reshape(-1, 1)).flatten()

# 2. MLP Neural Network
print("📌 Training MLP Neural Network...")
mlp_model = MLPRegressor(
    hidden_layer_sizes=(100, 50, 25),
    max_iter=500,
    random_state=42,
    early_stopping=True
)
mlp_model.fit(X_train, y_train)
mlp_pred_scaled = mlp_model.predict(X_test)
mlp_pred = scaler.inverse_transform(mlp_pred_scaled.reshape(-1, 1)).flatten()

# 3. Ridge Regression
print("📌 Training Ridge Regression...")
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
ridge_pred_scaled = ridge_model.predict(X_test)
ridge_pred = scaler.inverse_transform(ridge_pred_scaled.reshape(-1, 1)).flatten()

# Create DataFrames
ml_actual = test_data.values[window_size:]
ml_index = test_data.index[window_size:]

rf_predictions = pd.DataFrame({'Actual': ml_actual, 'Predicted': rf_pred}, index=ml_index)
mlp_predictions = pd.DataFrame({'Actual': ml_actual, 'Predicted': mlp_pred}, index=ml_index)
ridge_predictions = pd.DataFrame({'Actual': ml_actual, 'Predicted': ridge_pred}, index=ml_index)

print("\n✅ All ML models trained successfully!")
print("   - Random Forest")
print("   - MLP Neural Network")
print("   - Ridge Regression")
```

```python
# Cell 9: Model Evaluation (Fixed - No LSTM)

print("="*80)
print("📊 MODEL PERFORMANCE EVALUATION")
print("="*80)

def calculate_metrics(actual, predicted, model_name):
    """
    Calculate and return evaluation metrics.
    """
    # Ensure arrays are 1D
    actual = np.array(actual).flatten()
    predicted = np.array(predicted).flatten()
    
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    
    return {
        'Model': model_name,
        'MAE': mae,
        'RMSE': rmse,
        'MAPE': mape
    }

# Initialize metrics list
metrics_list = []

# ============================================================
# 1. ARIMA Metrics
# ============================================================
try:
    arima_actual = arima_predictions['Actual'].values
    arima_pred = arima_predictions['Predicted'].values
    arima_metrics = calculate_metrics(arima_actual, arima_pred, 'ARIMA')
    metrics_list.append(arima_metrics)
    print("✅ ARIMA metrics calculated")
except NameError:
    print("⚠️ ARIMA predictions not found - skipping")

# ============================================================
# 2. Random Forest Metrics
# ============================================================
try:
    rf_actual = rf_predictions['Actual'].values
    rf_pred = rf_predictions['Predicted'].values
    rf_metrics = calculate_metrics(rf_actual, rf_pred, 'Random Forest')
    metrics_list.append(rf_metrics)
    print("✅ Random Forest metrics calculated")
except NameError:
    print("⚠️ Random Forest predictions not found - skipping")

# ============================================================
# 3. MLP Neural Network Metrics
# ============================================================
try:
    mlp_actual = mlp_predictions['Actual'].values
    mlp_pred = mlp_predictions['Predicted'].values
    mlp_metrics = calculate_metrics(mlp_actual, mlp_pred, 'MLP Neural Net')
    metrics_list.append(mlp_metrics)
    print("✅ MLP metrics calculated")
except NameError:
    print("⚠️ MLP predictions not found - skipping")

# ============================================================
# 4. Ridge Regression Metrics
# ============================================================
try:
    ridge_actual = ridge_predictions['Actual'].values
    ridge_pred = ridge_predictions['Predicted'].values
    ridge_metrics = calculate_metrics(ridge_actual, ridge_pred, 'Ridge Regression')
    metrics_list.append(ridge_metrics)
    print("✅ Ridge metrics calculated")
except NameError:
    print("⚠️ Ridge predictions not found - skipping")

# ============================================================
# 5. LSTM Metrics (if available)
# ============================================================
try:
    lstm_actual = lstm_predictions['Actual'].values
    lstm_pred = lstm_predictions['Predicted'].values
    lstm_metrics = calculate_metrics(lstm_actual, lstm_pred, 'LSTM')
    metrics_list.append(lstm_metrics)
    print("✅ LSTM metrics calculated")
except NameError:
    print("⚠️ LSTM predictions not found - skipping (TensorFlow not available)")

# ============================================================
# 6. Create Comparison DataFrame
# ============================================================
if len(metrics_list) == 0:
    print("\n❌ No metrics available! Please run model training cells first.")
else:
    comparison_df = pd.DataFrame(metrics_list).round(2)

    print("\n" + "="*80)
    print("📊 Model Performance Comparison")
    print("="*80)
    display(comparison_df)

    # ============================================================
    # 7. Visualize Performance
    # ============================================================
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle('Model Performance Comparison', fontsize=16)

    metrics_to_plot = ['MAE', 'RMSE', 'MAPE']
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

    for idx, metric in enumerate(metrics_to_plot):
        axes[idx].bar(comparison_df['Model'], comparison_df[metric], color=colors[idx])
        axes[idx].set_title(f'{metric} Comparison', fontsize=12)
        axes[idx].set_ylabel(metric)
        axes[idx].grid(True, alpha=0.3)
        
        # Add value labels on bars
        for i, v in enumerate(comparison_df[metric]):
            axes[idx].text(i, v + 0.05 * v, f'{v:.2f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

    # ============================================================
    # 8. Best Model Analysis
    # ============================================================
    print("\n" + "="*80)
    print("🏆 BEST PERFORMING MODEL")
    print("="*80)

    best_idx = comparison_df['RMSE'].idxmin()
    best_model = comparison_df.loc[best_idx, 'Model']
    best_rmse = comparison_df.loc[best_idx, 'RMSE']

    print(f"   Best Model: {best_model}")
    print(f"   RMSE: ${best_rmse:.2f}")
    print(f"   MAE: ${comparison_df.loc[best_idx, 'MAE']:.2f}")
    print(f"   MAPE: {comparison_df.loc[best_idx, 'MAPE']:.2f}%")

    # ============================================================
    # 9. Summary
    # ============================================================
    print("\n" + "="*80)
    print("📌 MODEL SUMMARY")
    print("="*80)
    print(f"   Models Evaluated: {', '.join(comparison_df['Model'].tolist())}")
    print(f"   LSTM Status: ❌ Not available (TensorFlow DLL issue)")
    print(f"   Best Model: {best_model}")

    print("\n✅ Model evaluation complete! Ready for Task 3.")
    print("="*80)
```
